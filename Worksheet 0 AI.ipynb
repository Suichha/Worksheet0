```python
def sum_nested_list(nested_list):
    """
    Recursively calculates the sum of all numbers in a nested list.

    Args:
        nested_list (list): A list containing integers or sublists.

    Returns:
        int: Total sum of all numbers.
    """
    total = 0
    for element in nested_list:
        if isinstance(element, list):       
            total += sum_nested_list(element)
        else:
            total += element                
    return total



nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print("Sum of nested list =", sum_nested_list(nested_list))
```

```python
def generate_permutations(s):
    """
    Recursively generates all unique permutations of a string.

    Args:
        s (str): Input string.

    Returns:
        list: List of unique permutations.
    """
  
    if len(s) == 1:
        return [s]

    permutations = []

    for i in range(len(s)):
        current_char = s[i]
        remaining = s[:i] + s[i+1:]

        for perm in generate_permutations(remaining):
            new_perm = current_char + perm
            if new_perm not in permutations:
                permutations.append(new_perm)

    return permutations



print("Permutations of 'abc':", generate_permutations("abc"))
print("Permutations of 'aab':", generate_permutations("aab"))
```

```python
def calculate_directory_size(directory):
    """
    Recursively calculates the total size of a directory.

    Args:
        directory (dict): Keys = file/directory names,
                          Values = file size or nested dictionary.

    Returns:
        int: Total size in KB.
    """
    total_size = 0

    for key, value in directory.items():
        if isinstance(value, dict):
         
            total_size += calculate_directory_size(value)
        else:
            
            total_size += value

    return total_size


# Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}


print("Total directory size =", calculate_directory_size(directory_structure), "KB")
```

```python
def min_coins(coins, amount):
    """
    Finds the minimum number of coins needed to make up a given amount using DP (Tabulation).

    Parameters:
        coins (list): Available coin denominations.
        amount (int): Target amount.

    Returns:
        int: Minimum number of coins, or -1 if not possible.
    """

    dp = [float("inf")] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float("inf") else -1



coins = [1, 2, 5]
amount = 11
print("Minimum coins required:", min_coins(coins, amount))  
```

```python
def longest_common_subsequence(s1, s2):
    """
    Computes the length of the Longest Common Subsequence (LCS)
    using Dynamic Programming (Tabulation).

    Parameters:
        s1, s2 (str): Input strings.

    Returns:
        int: Length of the LCS.
    """

    n, m = len(s1), len(s2)

    # DP table (n+1 × m+1)
    dp = [[0] * (m + 1) for _ in range(n + 1)]

  
    for i in range(1, n + 1):
        for j in range(1, m + 1):

            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = 1 + dp[i - 1][j - 1]
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[n][m]



s1 = "abcde"
s2 = "ace"
print("LCS length:", longest_common_subsequence(s1, s2))  
```

```python
def knapsack(weights, values, capacity):
    """
    Solves the 0/1 Knapsack problem using Dynamic Programming.

    Parameters:
        weights (list): Weight of each item.
        values (list): Value of each item.
        capacity (int): Maximum weight capacity.

    Returns:
        int: Maximum total value that fits in the knapsack.
    """

    n = len(weights)

    # DP table (n+1 × capacity+1)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

   
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):

            if weights[i - 1] <= w:
           
                dp[i][w] = max(
                    values[i - 1] + dp[i - 1][w - weights[i - 1]],
                    dp[i - 1][w]
                )
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]


weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

print("Maximum knapsack value:", knapsack(weights, values, capacity))  
```
